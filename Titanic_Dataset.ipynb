```python
# import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load a sample dataset

# Load the Titanic dataset from seaborn
titanic = sns.load_dataset('titanic')
print("Sample of the Titanic dataset:")
print(titanic.head())

# Select only numeric columns for visualization
titanic = titanic[['age', 'fare', 'pclass', 'survived']]

# Drop missing values
titanic = titanic.dropna()

# 2. Plot the distribution of a feature

# Plot histogram of 'age'
plt.figure(figsize=(8, 6))
plt.hist(titanic['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Create scatter plots to understand relationships between features
# Scatter plot for 'age' vs 'fare'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', hue='pclass', data=titanic, palette='viridis')
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# 4. Use a correlation heatmap to find relationships between multiple features

# Compute the correlation matrix
correlation_matrix = titanic.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap (Titanic Dataset)')
plt.show()
```
