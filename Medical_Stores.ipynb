```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# -------------------------------
# DATA (Replace with Table 7 values)
# -------------------------------
# X = Number of competitor medical stores
# Y = Daily sales volume

X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)   # example data
Y = np.array([500, 480, 450, 420, 390, 360])     # example data

# -------------------------------
# SCATTER DIAGRAM
# -------------------------------
plt.scatter(X, Y)
plt.xlabel("Number of Competitor Medical Stores (within 1 km)")
plt.ylabel("Daily Sales Volume")
plt.title("Scatter Diagram: Competitors vs Sales")
plt.show()

# -------------------------------
# LINEAR REGRESSION MODEL
# -------------------------------
model = LinearRegression()
model.fit(X, Y)

# Regression coefficients
a = model.intercept_
b = model.coef_[0]

print("Linear Regression Model:")
print(f"Sales = {a:.2f} + {b:.2f} Ã— Competitors")

# -------------------------------
# REGRESSION LINE
# -------------------------------
Y_pred = model.predict(X)

plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("Number of Competitor Medical Stores (within 1 km)")
plt.ylabel("Daily Sales Volume")
plt.title("Linear Regression Model")
plt.show()
```
