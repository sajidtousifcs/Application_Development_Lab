```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Data from the table
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Oct", "Nov", "Dec"],
    "Actual": [112, 113, 114, 115, 112, 121, 122, 114],
    "Predicted": [113, 112, 116, 117, 110, 118, 121, 115]
}

# Create DataFrame
df = pd.DataFrame(data)
print("Monthly Sales Data (in Lakhs):\n")
print(df)

# Error Metrics
mae = mean_absolute_error(df["Actual"], df["Predicted"])
mse = mean_squared_error(df["Actual"], df["Predicted"])
rmse = np.sqrt(mse)

print("\nError Metrics:")
print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)

# Plot Actual vs Predicted Sales
plt.figure(figsize=(8, 5))
plt.plot(df["Month"], df["Actual"], marker='o', label="Actual Sales")
plt.plot(df["Month"], df["Predicted"], marker='s', linestyle='--', label="Predicted Sales")
plt.xlabel("Month")
plt.ylabel("Sales (Lakhs INR)")
plt.title("Actual vs Predicted Monthly Sales")
plt.legend()
plt.grid(True)
plt.show()
```
