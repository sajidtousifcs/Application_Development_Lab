```python
import numpy as np
from sklearn.linear_model import LinearRegression

# -------------------------------
# GIVEN DATA
# -------------------------------
# Age (x)
X = np.array([43, 21, 25, 42, 57, 59]).reshape(-1, 1)

# Glucose Level (y)
Y = np.array([99, 65, 79, 75, 87, 81])

# -------------------------------
# LINEAR REGRESSION MODEL
# -------------------------------
model = LinearRegression()
model.fit(X, Y)

# Predict glucose level for age = 49
age_to_predict = np.array([[49]])
predicted_glucose = model.predict(age_to_predict)

print("Predicted Glucose Level for age 49:")
print(round(predicted_glucose[0], 2))
```
