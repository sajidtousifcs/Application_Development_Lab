```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# -------------------------------
# DATA FROM TABLE 7
# -------------------------------
# Number of competitors (X)
X = np.array([1, 1, 2, 3, 3, 5, 5, 6]).reshape(-1, 1)

# Sales volume (Y)
Y = np.array([3600, 3300, 3100, 2900, 2700, 2300, 2000, 1800])

# -------------------------------
# SCATTER DIAGRAM
# -------------------------------
plt.scatter(X, Y)
plt.xlabel("Number of Competitor Medical Stores (within 1 km)")
plt.ylabel("Daily Sales Volume")
plt.title("Scatter Diagram: Competitors vs Sales")
plt.show()

# -------------------------------
# LINEAR REGRESSION MODEL
# -------------------------------
model = LinearRegression()
model.fit(X, Y)

# Regression coefficients
a = model.intercept_
b = model.coef_[0]

print("Linear Regression Equation:")
print(f"Sales = {a:.2f} + {b:.2f} Ã— Competitors")

# -------------------------------
# REGRESSION LINE
# -------------------------------
Y_pred = model.predict(X)

plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("Number of Competitor Medical Stores (within 1 km)")
plt.ylabel("Daily Sales Volume")
plt.title("Linear Regression Line")
plt.show()
```
